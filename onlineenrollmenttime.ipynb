```stata
use data/ipeds/hdefa_dist20122019.dta, clear
```

```stata
qui {
    drop if control<1
replace control=4 if iclevel>1 & control==1
collapse (sum) efdeexcg efdeexcu efdeexcd, by(year control)
reshape wide efdeexcg efdeexcu efdeexcd, i(year) j(control)
label var efdeexcg4 "community     "
label var efdeexcg3 "for-profit    "
label var efdeexcg2 "private"
label var efdeexcg1 "state"  
label var efdeexcu4 "community     "
label var efdeexcu3 "for-profit    "
label var efdeexcu2 "private"
label var efdeexcu1 "state"  
label var efdeexcd4 "community     "
label var efdeexcd3 "for-profit    "
label var efdeexcd2 "private"
label var efdeexcd1 "state"  
}
```

```stata
tw (connect efdeexcd* year if year>2011 & year<2020, sort ///
legend(pos(3) size(large) order(1 4 2 3)) xlabel(2010(2)2020, ang(45) labsize(medium)) ///
ytitle("exclusively online enrollment" " " " " "  ", size(large)) ylabel(, format(%10.0fc) labsize(medium)) ///
xtitle("") title({bf:Figure 1:} exclusively online enrollment - all levels, ///
size(large)) note("") scheme(plotplain)) 
```

```stata
tw (connect efdeexcu* year if year>2011 & year<2020, sort ///
legend(pos(3) size(large) order(4 1 2 3)) xlabel(2010(2)2020, ang(45) labsize(medium)) ///
ytitle("exclusively online enrollment" " " " " "  ", size(large)) ylabel(, format(%10.0fc) labsize(medium)) ///
xtitle("") title({bf:Figure 2:} exclusively online enrollment - undergrad, ///
size(large)) note("") scheme(plotplain)) 
```

```stata
tw (connect efdeexcg1 efdeexcg2 efdeexcg3  year if year>2011 & year<2020, sort ///
legend(pos(3) size(large)) xlabel(2010(2)2020, ang(45) labsize(medium)) ///
ytitle("exclusively online enrollment" " " " " "  ", size(large)) ylabel(, format(%10.0fc) labsize(medium)) ///
xtitle("") title({bf:Figure 3:} exclusively online enrollment - grad, ///
size(large)) note("") scheme(plotplain)) 
```
